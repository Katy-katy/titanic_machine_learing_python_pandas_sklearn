```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
data = sns.load_dataset('titanic') # we can download this dataset form Seaborn directly 
data.head()
```

Let's compare the number of femals vs. males

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=data)
```

```python
g = sns.FacetGrid(data=data,col='sex')
g.map(plt.hist,'age')
```

Let's find the features that are most correlated with "survived" using heatmap

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True)
```

We can see that "aduilt_male" has strong negative correlation. "Pclass" and "alone" have also negative correlation ("Pclass" is 1,2,3 and 3 is this cheapest one). "Fare" have positive correlation. 

This vizualiztion also shows that some fatures are hitghly correlaed each other. For example: "adult_male" and "alone", "sibsp" and "parch". We should remember about this when we will use machine learining algorithms. 

```python
sns.jointplot(x='fare',y='age',data=data,kind='scatter')
```

```python
plt.figure(figsize=(12,5))
sns.distplot(data['fare'], kde=False, bins=50)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='class',data=data, palette='rainbow')
```

```python
sns.boxplot(x="class", y="age", data=data,palette='rainbow')
```

```python
plt.figure(figsize=(10,4))
sns.swarmplot(x="class", y="age", data=data, split=True, hue='sex',palette='Set1')
```

```python
plt.figure(figsize=(10,4))
sns.swarmplot(x="class", y="age", data=data, split=True, hue='survived',palette='Set1')
```

This visualization shows that almost all kids form firs and the second class survived and that passengers from the first class had much more chance to survived that from third class.
